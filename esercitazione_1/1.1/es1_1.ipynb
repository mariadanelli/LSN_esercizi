```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)

# Carica i dati dal file
data = np.loadtxt('media.dat')

x*=1000

plt.errorbar(x,data[:,0],yerr=data[:,1])
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

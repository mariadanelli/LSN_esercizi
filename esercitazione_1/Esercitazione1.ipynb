### Esercizio 1.1

Nella seguente esercitazione andremo a stimare la media e l'incertezza di una variabile casuale uniformemente distribuita fra $[0,1] $, eseguiremo questa stima con il metodo della media a blocchi, scegliendo di fare $M_{throws}= 10^6$ totali divisi in $N_{blocchi}=100 $. 
Il metodo della media a blocchi consiste nell'eseguire la media all'interno del primo blocco, il risultato costituirà la nostra prima stima, il secondo risultato invece sarà costituito dalla media fra la media del secondo blocco e la media del primo blocco e così via.  L'incertezza è calcolata come $\sigma = \sqrt(\frac{\langle x^2 \rangle -  (\langle x \rangle)^2} {N} )$, dove $\langle x \rangle$ rappresenta la media eseguita fra le medie dei blocchi precedenti.
Il valore di aspettazione con cui andiamo a confrontare i nostri risultati è $\langle r \rangle = \int_0^1 r dr = 1/2$.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)

# Carica i dati dal file
data = np.loadtxt("1.1/media.dat")

x*=1000

plt.errorbar(x,data[:,0],yerr=data[:,1])
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

Ripetiamo la stessa procedura per stimare la varianza, il cui valore di aspettazione è invece $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)

# Carica i dati dal file
data = np.loadtxt('1.1/varianza.dat')

x*=1000

plt.errorbar(x,data[:,0],yerr=data[:,1])
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

Per completare la nostra analisi statisica eseguiamo il Test del $\chi ^2$ : tramite questo possiamo accettare o rifiutare l'ipotesi che i nostri dati NON provengano da una distribuzione uniforme, ci apsettiamo  $\chi^2 \simeq 100$ dato che $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$ ($M=100$) e dato che, se i nostri risultati si distribuiscono secondo una distribuzione uniforme,  $(n_i - n/M)^2 \simeq n/M$


```python
import matplotlib.pyplot as plt
import numpy as np
import statistics

# Carica i dati dal file
x= np.arange(100)
data = np.loadtxt('1.1/chiq.dat')


plt.bar(x, data)
plt.xlabel('j')
plt.ylabel('$\chi ^2$')
plt.title('Test del $\chi ^2$')
plt.grid(True)
print('Media:',statistics.mean(data))
print('Deviazione Standard :',np.round((statistics.variance(data))**0.5,4))
plt.show()
```

È stato calcolato $S_N$ $10^4$ volte, con $x_i$ estratto da una distribuzione

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.optimize import curve_fit

# Carica i dati dal file
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 5))

data1 = np.loadtxt('1.2/cento.txt')
hist1, bins1, _ = ax1.hist(data1, 100, density=True, color='green', label='Uniforme')
ax1.set_xlabel ('random x')
ax1.set_ylabel ('probabilità')

data2 = np.loadtxt('1.2/cento exp.txt')
hist2, bins2, _ = ax2.hist(data2, 100, density=True, color='red', label='Esponenziale')
ax2.set_xlabel ('random x')
ax2.set_ylabel ('probabilità')

data3 = np.loadtxt('1.2/cento lor.txt')
ax3.hist(data3, 100, (-30,30) ,density=True,color='blue', label= 'Cauchy Lorentz')
ax3.set_xlabel ('random x')
ax3.set_ylabel ('probabilità')

def gaussian_function(x, amplitude, mean, std_dev):
    return amplitude * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2))

x = np.linspace(0, 1, 10000)

bin_centers1 = 0.5 * (bins1[:-1] + bins1[1:])
params, covariance = curve_fit(gaussian_function, bin_centers1, hist1, p0=[1, np.mean(data1), np.std(data1)])
# Estrai i parametri del fit
amplitude_fit, mean_fit, std_dev_fit = params

x= np.linspace( mean_fit-3*std_dev_fit, mean_fit+3*std_dev_fit)
y= gaussian_function(x, amplitude_fit, mean_fit, std_dev_fit)
ax1.plot(x,y, 'k--',label='fit')
ax1.legend()
ax1.grid(True)
x = np.linspace(0, 1, 10000)

bin_centers2= 0.5 * (bins2[:-1] + bins2[1:])
params, covariance = curve_fit(gaussian_function, bin_centers2, hist2, p0=[1, np.mean(data2), np.std(data2)])
# Estrai i parametri del fit
amplitude_fit, mean_fit, std_dev_fit = params

x= np.linspace( mean_fit-3*std_dev_fit, mean_fit+3*std_dev_fit)
y= gaussian_function(x, amplitude_fit, mean_fit, std_dev_fit)
ax2.plot(x,y, 'k--',label='fit')
ax2.legend()
ax2.grid(True)
fig.suptitle('Media su 100 elemento di una distribuzione di probabilità')
ax3.grid(True)
plt.show()
```

## Esercizio 1.3


Al fine di stimare il valore del numero irrazionale $ \pi $ simuliamo l'esperimento di Buffon:

 Una sbarra di lunghezza $L $ viene lanciata a caso su un piano orizzontale dove sono poste delle rette a una distanza $ (d)$ ,dove $(d > L). $ 
 La probabilità $P$ che la sbarra intersechi una di queste linee è data da: $P = \frac{2L}{\pi d} $. Possiamo quindi utilizzare ciò per stimare $\pi\ $ tramite i lanci della sbarra: se la sbarra viene lanciata $N_{throw}$  volte e si osserva che atterra su una linea $N_{hit}$ volte, allora:
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$


```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(100)

# Carica i dati dal file
data = np.loadtxt('1.3/media.dat')

x*=1000

plt.errorbar(x,data[:,0],yerr=data[:,1], label= 'simulazione')
y= np.full(100, math.pi)
plt.plot(x, y, '--' ,label= 'valore aspettato')
plt.xlabel('#throws')
plt.ylabel('$ \pi $')
plt.title ('Stima del valore di $\pi$')
plt.grid(True)
plt.legend()
plt.show()
```

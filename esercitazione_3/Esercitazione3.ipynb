Nella seguente esercitazione si vuole stimare il prezzo opportuno da assegnare a un'opzione. Per far ciò innazitutto consideriamo il profitto che l'holder dell'opzione fa al tempo T (delivery time) = $max[S(T)-K,0]$ per la call-option e = $max[K-S(0), 0]$ per la put-option, dove K è il prezzo fissato e $S(T)$ è il prezzo di mercato. Questo profitto va poi scontato di $e^{-rT}$ poichè dato un deposito $S(0)$ in un tempo T è garantito un profitto di $S(0)e^{rT}$.

 Il problema risiede nello stimare $S(T)=S(0) e^{((r-\frac{\sigma^{2}}{2}T+\sigma z \sqrt(T))};$ dove ${\sigma}$ è la volatilità, r è un tasso di interesse senza rischi, e z è una variabile casuale che segue una distribuzione gaussiana, stiamo in questo modo seguendo la teoria di Black-Scholes che assume che l'evoluzione temporale del prezzo sia stocastica e segua un moto Browniano.

Per fare questa stima si è eseguita una simulazione Montecarlo di $10^6 $ lanci, eseguendo l'analisi statistica con il metodo della media a blocchi, $S(T)$ è stato stimato
1. in maniera diretta
2. suddividendo l'intervallo di tempo in dei sottointervalli in ognuno dei quali $S(T)$ evolve in maniera casuale

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)

x*=10000

# Carica i dati dal file
fig, (ax, axD) = plt.subplots(1, 2, figsize=(16, 5))


# Carica i dati dal file
data = np.loadtxt('direttoP.dat')
ax.errorbar(x,data[:,0],yerr=data[:,1])
p= np.full(100,5.4595325819072364 )
ax.plot(x,p, label= 'valore atteso')
ax.set_title("Diretto",fontsize=16)
ax.set_xlabel('#throws')
ax.set_ylabel('Put')
ax.grid(True)
ax.legend()

dataD = np.loadtxt('discretizzatoP.dat')
axD.errorbar(x,dataD[:,0],yerr=dataD[:,1])
axD.plot(x,p, label= 'valore atteso')
axD.set_title("Discretizzato",fontsize=16)
axD.set_xlabel('#throws')
axD.set_ylabel('Put')
axD.grid(True)
axD.legend()
plt.suptitle('Put option europea', fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)

x*=10000

# Carica i dati dal file
fig, (ax, axD) = plt.subplots(1, 2, figsize=(16, 5))


# Carica i dati dal file
data = np.loadtxt('direttoC.dat')
ax.errorbar(x,data[:,0],yerr=data[:,1])
c= np.full(100, 14.975790778311286)
ax.plot(x,c, label='soluzione analitica')
ax.set_title("Diretto",fontsize=16)
ax.set_xlabel('#throws')
ax.set_ylabel('Call')
ax.grid(True)

dataD = np.loadtxt('discretizzatoC.dat')
axD.errorbar(x,dataD[:,0],yerr=dataD[:,1])
axD.plot(x,c, label='soluzione analitica')
axD.set_title("Discretizzato",fontsize=16)
axD.set_xlabel('#throws')
axD.set_ylabel('Call-14.9757908')
axD.grid(True)
plt.suptitle('Call option europea', fontsize=20)
plt.show()
```

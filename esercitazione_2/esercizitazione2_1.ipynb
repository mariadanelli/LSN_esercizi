```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)

# Carica i dati dal file
data = np.loadtxt('unif.dat')

x*=1000

plt.errorbar(x,data[:,0]-1,yerr=data[:,1])
plt.xlabel('#throws')
plt.ylabel('Integrale-1')
plt.grid(True)
plt.show()
```

Risolviamo ora nuovamente il problema del commesso viaggiatore, aumentando la complessità del problema, portando a $N= 50$ il numero di città da visitare. Per aumentare la potenza di calcolo viene utilizzata la libreria MPI per il calcolo parallelo, nello specifico sono stati utilizzati 4 nodi: 
* in maniera indipendente: ogni nodo esegue l'algoritmo e giunge alla propria soluzione ottimale
* facendoli comunicare tra di loro: ogni $N_{migrazioni}= 40$ ogni nodo riceve il miglior individuo da un altro, in tal modo la convergenza dell'algoritmo è migliorata come si vede dai risultati sotto riportati.

Utilizziamo delle popolazioni da $N_{pop}=1000$ individui, che facciamo evolvere per $N_{gen} =250 $ generazioni. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fig, (ax) = plt.subplots(1, 1, figsize=(11, 8))
data = np.loadtxt('best11.dat')

ax.plot(data[:, 0], data[:, 1], "o", c='g', label="Coordinate città")
ax.set_title("Disposizione delle capitali americane", fontsize=19)
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.set_aspect('equal') 
ax.grid(True)
ax.legend()
plt.show()
```
